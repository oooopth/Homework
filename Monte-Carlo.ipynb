```python
from math import sqrt
import matplotlib.pyplot as plt 
from matplotlib import mlab 


G = 6.67384 * 10 **(-11)
M1 = 400000
dM1 = 500
M2 = 300000
dM2 = 1000
R = 3.2
dR = 0.01
F = G * M1 * M2 / R**2
dF = F * sqrt((dM1 / M1) ** 2 + (2 * dR / R) ** 2 + (dM2 / M2) ** 2)

import numpy as np
n = 1000000
m1 = np.random.normal(M1, dM1, n)
m2 = np.random.normal(M2, dM2, n)
r = np.random.normal(R, dR, n)
f = G * m1 * m2 / r**2

plt.figure()
plt.hist(f, normed = True, bins = 30)
plt.grid(True)
x = np.linspace(F - 3 * dF, F + 3 * dF, n)
plt.plot(x, mlab.normpdf(x, F, dF))
plt.show()
```
