```python
m = numpy.zeros((8, 8))
for i in range(0, 8):
    for j in range(0, 8):
        if (i + j) % 2 == 1:
            m[i, j] = 1
print(m)
```

```python
import math
M = (numpy.random.rand(10, 2) - 0.5) * 10000
for i in range(0, 10):
    x = M[i, 0]
    y = M[i, 1]
    r = math.sqrt(x**2 + y**2)
    phi = numpy.math.atan2(y, x)
    M[i, 0] = r
    M[i, 1] = phi
print(M)
```
