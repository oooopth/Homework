```python
import numpy

def solve(M):                #Фукция, которая решает матрицу
    n = M.shape[0]
    for i in range(0, n - 1):
        if M[i + 1, i] != 0:  
            if M[i, i] != 0:
                M[i, :] *= M[i + 1, i] / M[i, i]
                M[i + 1, :] -= M[i, :]
            else:
                M[i, :], M[i + 1, :] = M[i + 1, :], M[i, :]
    x = numpy.zeros(n)
    x[n - 1] = M[n - 1, n] / M[n - 1, n - 1]
    for i in range(n - 2, -1, -1):
        x[i] = (M[i, n] - x[i + 1] * M[i, i + 1]) / M[i, i]
    return x
```

```python
from random import randint

def make_matrix():    #Рандомно задает матрицу
    n = 20
    left = -10000
    right = 10000
    M = numpy.zeros((n, n + 1))
    M[0, 0] = randint(left, right)
    M[0, 1] = randint(left, right)
    M[0, -1] = randint(left, right)
    for i in range(1, n):
        M[i, i - 1] = randint(left, right)
        M[i, i] = randint(left, right)
        M[i, i + 1] = randint(left, right)
        M[i, -1] = randint(left, right)
    return M
```

```python
import time

t0 = time.time()    #Считает время, за которое функция решает 100000 матриц
res = 0
for i in range(100000):  
    M = make_matrix()
    M0 = M[:, :-1].copy()
    b = M[:, -1].copy()
    x = solve(M)
    if (M0 @ x - b < 0.0001).all():   #Считает, сколько матриц было решено правильно из 100000
        res += 1
print(res)
print(time.time() - t0)
```
